```python
import pandas as pd
import matplotlib.pyplot as plt

# Data perusahaan
data = {
    'ID_Perusahaan': [1, 2, 3, 4, 5, 6, 7],
    'Nama_Perusahaan': [
        'PT.Pertamina Indonesia', 'PT Chevron Pasifik Indonesia', 'ExxonMobil Cepu Ltd',
        'Total E&P Indonesia', 'Petronas Carigali', 'Shell Indonesia', 'BP Berau Ltd'
    ],
    'Tahun_Pendirian': [1957, 1924, 2000, 1968, 1974, 2001, 2009]
}

df = pd.DataFrame(data)

# Urutkan dari perusahaan tertua ke termuda
df_sorted = df.sort_values(by='Tahun_Pendirian')

# Buat bar chart horizontal
plt.figure(figsize=(10, 6))
bars = plt.barh(df_sorted['Nama_Perusahaan'], df_sorted['Tahun_Pendirian'], color='skyblue')
plt.xlabel('Tahun Pendirian')
plt.title('Visualisasi Tahun Pendirian Perusahaan Minyak')

# Tambahkan label tahun di setiap bar
for bar in bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, str(int(bar.get_width())), va='center')

plt.tight_layout()
plt.gca().invert_yaxis()  # Supaya yang tertua di atas
plt.show()
```
