```python
# üöÄ Import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------------------
# üßπ 1Ô∏è‚É£ Data Cleaning & Preparation
df = pd.read_csv('/content/Produksi Jagung 2025.csv')

df_clean = df.iloc[3:].copy()
df_clean.columns = [
    'Provinsi', 'Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli',
    'Agustus', 'September', 'Oktober', 'November', 'Desember', 'Tahunan'
]
df_clean.reset_index(drop=True, inplace=True)
df_clean.replace('-', 0, inplace=True)

for col in df_clean.columns[1:]:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce').fillna(0)

# ---------------------------------------------------------------
# üìä 2Ô∏è‚É£ Statistik Deskriptif
print("üìä Statistik Deskriptif: Produksi Rata-rata Januari‚ÄìJuli per Provinsi")
df_clean['Rata2_JanJul'] = df_clean[['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli']].mean(axis=1)
display(df_clean[['Provinsi', 'Rata2_JanJul']])

# ---------------------------------------------------------------
# üìà 3Ô∏è‚É£ Prediksi Produksi Agustus
df_clean['Prediksi_Agustus'] = df_clean['Rata2_JanJul']
df_clean['Realisasi_Agustus'] = df_clean['Agustus']
df_clean['Selisih'] = df_clean['Realisasi_Agustus'] - df_clean['Prediksi_Agustus']
df_clean['Lonjakan(%)'] = (df_clean['Selisih'] / df_clean['Prediksi_Agustus']) * 100
df_clean['Status'] = df_clean['Selisih'].apply(lambda x: 'Lonjakan Positif' if x > 0 else 'Lonjakan Negatif')

# ---------------------------------------------------------------
# üîç 4Ô∏è‚É£ Analisis Perbandingan
print("üîç Tabel Hasil Perbandingan Prediksi vs Realisasi Agustus")
hasil = df_clean[['Provinsi', 'Prediksi_Agustus', 'Realisasi_Agustus', 'Selisih', 'Lonjakan(%)', 'Status']]
pd.set_option('display.float_format', '{:,.2f}'.format)
display(hasil)

# Provinsi dengan lonjakan signifikan (>20% atau <-20%)
print("\nüìà Provinsi dengan lonjakan positif signifikan (>20%):")
print(hasil[hasil['Lonjakan(%)'] > 20][['Provinsi', 'Lonjakan(%)']])

print("\nüìâ Provinsi dengan penurunan signifikan (<-20%):")
print(hasil[hasil['Lonjakan(%)'] < -20][['Provinsi', 'Lonjakan(%)']])

# ---------------------------------------------------------------
# üìä 5Ô∏è‚É£ Visualisasi Lonjakan Persentase Agustus
plt.figure(figsize=(12,6))
colors = df_clean['Status'].map({'Lonjakan Positif':'green', 'Lonjakan Negatif':'red'})
plt.bar(df_clean['Provinsi'], df_clean['Lonjakan(%)'], color=colors)
plt.axhline(0, color='black', linestyle='--')
plt.xticks(rotation=90)
plt.title('Lonjakan (%) Produksi Agustus per Provinsi')
plt.ylabel('Lonjakan (%)')
plt.tight_layout()
plt.show()

# ---------------------------------------------------------------
# üìä 6Ô∏è‚É£ Analisis Tren Musiman Nasional
bulan = [
    'Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus',
    'September', 'Oktober', 'November', 'Desember'
]
nasional_bulanan = df_clean[bulan].sum()

plt.figure(figsize=(10,6))
plt.plot(bulan, nasional_bulanan, marker='o', color='blue')
plt.title('Tren Musiman Produksi Jagung Nasional 2025')
plt.ylabel('Produksi Nasional (ton)')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()

# ---------------------------------------------------------------
# üìù 7Ô∏è‚É£ Simpan Hasil Analisis ke File CSV
hasil.to_csv('/content/Hasil_Analisis_Produksi_Jagung_2025.csv', index=False)
print("‚úÖ Hasil analisis disimpan ke: Hasil_Analisis_Produksi_Jagung_2025.csv")
```

```python
# üöÄ Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# =============================================
# üìä 1Ô∏è‚É£ Exploratory Data Analysis (EDA)
# =============================================

# Baca data
df = pd.read_csv('/content/Produksi Jagung 2025.csv')

# Intip data semua provinsi (39 provinsi)
print("üìÑ Data 39 provinsi:")
display(df.head(39))

# Bersihkan header tambahan
df_clean = df.iloc[3:].copy()
df_clean.columns = [
    'Provinsi', 'Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli',
    'Agustus', 'September', 'Oktober', 'November', 'Desember', 'Tahunan'
]
df_clean.reset_index(drop=True, inplace=True)
df_clean.replace('-', 0, inplace=True)

# Konversi angka
for col in df_clean.columns[1:]:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce').fillna(0)

print("‚úÖ Data sudah dibersihkan. Dimensi:", df_clean.shape)

# Statistik deskriptif
print("üìä Statistik deskriptif:")
display(df_clean.describe())

# Korelasi antar bulan
bulan = [
    'Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus',
    'September', 'Oktober', 'November', 'Desember'
]
plt.figure(figsize=(10,8))
sns.heatmap(df_clean[bulan].corr(), annot=True, cmap='coolwarm')
plt.title("üìà Korelasi Produksi Bulanan per Provinsi")
plt.show()

# Tren musiman nasional
nasional_bulanan = df_clean[bulan].sum()
plt.figure(figsize=(10,6))
plt.plot(bulan, nasional_bulanan, marker='o', color='blue')
plt.title("üìà Tren Musiman Produksi Jagung Nasional 2025")
plt.ylabel("Produksi Nasional (ton)")
plt.xticks(rotation=45)
plt.grid()
plt.show()

# Tampilkan semua data 39 provinsi
print("üìÑ Data lengkap 39 provinsi setelah preprocessing:")
display(df_clean)

# =============================================
# üîç 2Ô∏è‚É£ Data Preprocessing
# =============================================

# Fitur & target
X = df_clean[['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli']]
y_actual = df_clean['Agustus']

# Normalisasi fitur
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# =============================================
# üìà 3Ô∏è‚É£ Modelling
# =============================================

print("‚úÖ Modelling: Latih model regresi linear untuk prediksi Agustus")

# Latih model regresi linear
model = LinearRegression()
model.fit(X_scaled, y_actual)

# Hitung prediksi Agustus
y_pred = model.predict(X_scaled)

# Masukkan hasil ke dataframe
df_clean['Prediksi_Agustus'] = y_pred
df_clean['Realisasi_Agustus'] = y_actual
df_clean['Selisih'] = df_clean['Realisasi_Agustus'] - df_clean['Prediksi_Agustus']
df_clean['Lonjakan(%)'] = (df_clean['Selisih'] / df_clean['Prediksi_Agustus']) * 100
df_clean['Status'] = df_clean['Selisih'].apply(lambda x: 'Lonjakan Positif' if x > 0 else 'Lonjakan Negatif')

# Tabel hasil
hasil = df_clean[['Provinsi', 'Prediksi_Agustus', 'Realisasi_Agustus', 'Selisih', 'Lonjakan(%)', 'Status']]
print("üìä Hasil prediksi, selisih, lonjakan (%) & status per provinsi:")
display(hasil)

# Visualisasi hasil lonjakan per provinsi
plt.figure(figsize=(14,7))
colors = hasil['Status'].map({'Lonjakan Positif':'green', 'Lonjakan Negatif':'red'})
plt.bar(hasil['Provinsi'], hasil['Lonjakan(%)'], color=colors)
plt.axhline(0, color='black', linestyle='--')
plt.xticks(rotation=90)
plt.title("üìà Lonjakan (%) Produksi Agustus per Provinsi (39 Provinsi)")
plt.ylabel("Lonjakan (%)")
plt.tight_layout()
plt.show()

# =============================================
# üîó Simpan hasil
hasil.to_csv('/content/Hasil_EDA_Prepr_Model_Produksi_Jagung.csv', index=False)
print("‚úÖ Hasil analisis & prediksi disimpan ke: Hasil_EDA_Prepr_Model_Produksi_Jagung.csv")
```
