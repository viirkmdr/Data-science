```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Buat DataFrame
data = {
    "Nama_Perusahaan": [
        "PT Pertamina Indonesia", "PT Chevron Pacific Indonesia", "ExxonMobil Cepu Ltd",
        "Total E&P Indonesie", "Petronas Carigali", "Shell Indonesia", "BP Berau Ltd"
    ],
    "Negara": ["Indonesia", "Indonesia", "Indonesia", "Indonesia", "Malaysia", "Indonesia", "Indonesia"],
    "Produksi_Harian_Barel": [210000, 170000, 200000, 160000, 90000, 120000, 110000]
}
df = pd.DataFrame(data)

# Visualisasi Produksi Harian
plt.figure(figsize=(10, 6))
sns.barplot(x="Produksi_Harian_Barel", y="Nama_Perusahaan", data=df, palette="Blues_d")
plt.title("Produksi Harian Minyak per Perusahaan")
plt.xlabel("Produksi Harian (barel)")
plt.ylabel("Perusahaan")
plt.show()
```

Berdasarkan visualisasi produksi harian perusahaan minyak diatas, terlihat bahwa PT Pertamina Indonesia merupakan produsen terbesar dengan estimasi produksi sekitar 210.000 barel per hari. Disusul oleh ExxonMobil Cepu Ltd yang memproduksi sekitar 200.000 barel per hari, dan PT Chevron Pacific Indonesia dengan sekitar 170.000 barel per hari. Rata-rata produksi harian dari tujuh perusahaan yang ditampilkan berada pada kisaran 151.429 barel per hari, dengan nilai median sebesar 160.000 barel per hari. Produksi terendah tercatat pada Petronas Carigali dengan sekitar 90.000 barel per hari, sementara produksi tertinggi oleh Pertamina menciptakan rentang (range) produksi sebesar 120.000 barel per hari.
